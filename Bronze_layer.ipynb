### **Dynamic Parameters**

```python
dbutils.widgets.text("file_name","")
p_file_name = dbutils.widgets.get("file_name")
```

### **Reading Data**

```python
df = spark.readStream.format("cloudFiles")\
        .option("cloudFiles.format","parquet")\
        .option("cloudFiles.schemaLocation",f"abfss://bronze@project1ds.dfs.core.windows.net/checkpoint_{p_file_name}")\
        .load(f"abfss://source@project1ds.dfs.core.windows.net/{p_file_name}")
```

### **Writing Data**

```python
df.writeStream.format("parquet")\
    .outputMode("append")\
    .option("checkpointLocation",f"abfss://bronze@project1ds.dfs.core.windows.net/checkpoint_{p_file_name}")\
    .option("path",f"abfss://bronze@project1ds.dfs.core.windows.net/{p_file_name}")\
    .trigger(once=True)\
    .start()
```

```python
df = spark.read.format("parquet")\
        .load(f"abfss://bronze@project1ds.dfs.core.windows.net/{p_file_name}")
        
display(df)
```
