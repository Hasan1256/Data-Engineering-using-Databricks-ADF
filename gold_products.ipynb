# **DLT Pipeline**

### **Expectations**

```python
my_rules = {
    "rule1": "product_id is NOT NULL",
    "rule2": "product_name is NOT NULL"
}
```

### **Streaming Table**

```python
import dlt
from pyspark.sql.functions import *
```

```python
@dlt.table()

@dlt.expect_all_or_drop(my_rules)

def DimProducts_stage():
    df = spark.readStream.table("project1_catalog.silver.products_silver")
    return df
```

### **Streaming View**

```python
@dlt.view

def DimProducts_view():
    df = spark.readStream.table("Live.DimProducts_stage")
    return df
```

### **DimProducts**

```python
dlt.create_streaming_table("DimProducts")
```

```python
dlt.apply_changes(
    target = "DimProducts",   # target Delta table
    source = "Live.DimProducts_view",               # source streaming table/view
    keys = ["product_id"],                # primary key(s) for matching
    sequence_by = "product_id",    # column to order changes (e.g., timestamp, version)
    stored_as_scd_type = 2            # choose SCD1 or SCD2
)
```
